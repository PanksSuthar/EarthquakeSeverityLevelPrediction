```python
import pandas as pd
import numpy as np
```

```python
train_df =pd.read_csv("train.csv")
train_df.head(10)
```

```python
def create_master_df():
    building_ownership_df =pd.read_csv("Building_Ownership_Use.csv")
    building_structure_df =pd.read_csv("Building_Structure.csv")
    temp_df = pd.merge(train_df,building_ownership_df, how="inner", on="building_id")
    master_df = pd.merge(temp_df,building_structure_df, how="inner", on="building_id")
    return master_df
```

```python
#master_df = create_master_df()
master_df = pd.read_csv("master.csv")
```

```python
temp=master_df
```

```python
temp =temp.drop(['vdcmun_id_x','district_id','district_id_x','district_id','vdcmun_id','ward_id_y','district_id_y','vdcmun_id_y','ward_id_x'], axis=1)
temp.head(5)
```

```python
# One Hot Encoding
temp = pd.get_dummies(temp, columns=['area_assesed','legal_ownership_status','position','land_surface_condition','foundation_type',
                                     'roof_type','ground_floor_type','other_floor_type','plan_configuration',
                                     'condition_post_eq'])

temp.info()
```

```python
temp = temp.loc[:, ~temp.columns.str.contains('^Unnamed')]
```

```python
temp2 = temp
```

```python
temp2 = temp2.replace([np.inf, -np.inf], np.nan).dropna(how="any")
```

```python
import keras
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
temp2 =temp2.drop(['building_id'],axis =1)
```

```python
y= temp2.iloc[:,0]
y = pd.get_dummies(y, columns=['damage_grade'])
x = temp2.iloc[:,1:]
x.head()
```

```python
# Splitting data

(X_train, X_test, Y_train, Y_test) = train_test_split(x, y, test_size=0.30, random_state=20)
df = np.where(np.isnan(Y_test))
```

```python
import h5py

model = keras.models.Sequential()

model.add(keras.layers.Dense(89, input_dim=89, init='uniform', activation='sigmoid'))
model.add(keras.layers.Dense(10, init='uniform', activation='sigmoid'))
model.add(keras.layers.Dense(10, init='uniform', activation='sigmoid'))
model.add(keras.layers.Dense(5, init='uniform', activation='sigmoid'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, Y_train, validation_data=(X_test, Y_test), nb_epoch=100, batch_size=1000)

scores = model.evaluate(X_test, Y_test)
print ("Accuracy: %.2f%%" %(scores[1]*100))

model.save('model.h5')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
#clf = tree.DecisionTreeClassifier()
clf = RandomForestClassifier(n_estimators = 10)
clf.fit(X_train,Y_train)
Y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))
```
